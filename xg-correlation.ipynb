```python
#  Ligaの2021/2022シーズンの全シュートデータを取得する
event_json_data_list = []
for match_id in ser_match.values:
    with open(f"./data/events/{match_id}.json") as datafile:
        data = json.load(datafile)
    event_json_data_list.append(data)
    
df_event_list = pd.concat(list(map(lambda x: pd.DataFrame(x), event_json_data_list)))
df_shot_list = df_event_list[df_event_list['shot'].notnull()]
df_shot_list.shape
```
